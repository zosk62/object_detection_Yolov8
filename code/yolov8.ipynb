```python
import cv2
from ultralytics import YOLO
import math
from pytube import YouTube

                                                                       
# yolo_model = YOLO("yolo-Weights/yolov8n.pt")
# yolo_model = YOLO('C:/Users/user/Data_Ana_3/Sprint2/runs/detect/yolov8sdrone.pt/weights/best.pt')
# yolo_model = YOLO('C:/Users/user/Data_Ana_3/Sprint2/runs/detect/yolov8sdronebird25.pt/weights/best.pt')
# yolo_model = YOLO('C:/Users/user/Data_Ana_3/Sprint2/runs/detect/train2/weights/best.pt')
yolo_model = YOLO('C:/Users/user/Data_Ana_3/Sprint2/Final_files/runs/detect/yolov8sdronebirdlight.pt/weights/best.pt')  # 70 times training
# yolo_model = YOLO('C:/Users/user/Data_Ana_3/Sprint2/runs/detect/yolov8sdronebird25.pt3/weights/best.pt')  # 35 times training other computer



# YOLOv8 classes
yolo_classes = yolo_model.names

# YouTube video URL
# video_url = "https://www.youtube.com/watch?v=PX9p7VES3bk"
# video_url = "https://www.youtube.com/watch?v=Hhp11I-vGHA" # drone
# video_url = "https://www.youtube.com/shorts/EohBaUH0Ycw" # bird
# video_url = "https://www.youtube.com/shorts/8A8bN-YgEpw" # bird
# video_url = "https://www.youtube.com/watch?v=a1wp1RnC7kk" 
# video_url = "https://www.youtube.com/watch?v=ZO5lV0gh5i4&t=91s"
# video_url = "https://www.youtube.com/watch?v=a1wp1RnC7kk&t=22s"
# video_url ="https://www.youtube.com/watch?v=a1wp1RnC7kk"
# video_url ="https://www.youtube.com/watch?v=Hhp11I-vGHA&t=23s"
video_url ="https://www.youtube.com/watch?v=Y1zhsObP49o"


# Extract video stream URL using pytube
yt = YouTube(video_url)
video_stream_url = yt.streams.filter(file_extension="mp4", res="720p").first().url

# Webcam configuration with the video stream URL
cap = cv2.VideoCapture(video_stream_url)

while True:
    # Capture frame from webcam
    success, img = cap.read()

    # Check if the frame is None
    if img is None:
        break

    # YOLOv8 detection on webcam frame
    results = yolo_model(img, stream=True)

    # Draw bounding boxes on webcam frame if confidence is more than 0.5
    for r in results:
        boxes = r.boxes
        for box in boxes:
            confidence = box.conf[0]
            if confidence > 0.5:
                x1, y1, x2, y2 = box.xyxy[0].int().tolist()
                cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 255), 3)

                # Display confidence score
                text = f"{yolo_classes[int(box.cls[0])]}: {confidence:.2f}"

                # Draw text near the bounding box with different color and thickness
                org_text = (x1, y1 - 10)  # Adjust the position of the text
                font = cv2.FONT_HERSHEY_SIMPLEX
                fontScale = 0.5  # Adjust the font size
                color = (255, 0, 0)  # Change BGR values for a different color (blue in this case)
                thickness = 2  # Increase thickness for a bolder font
                cv2.putText(img, text, org_text, font, fontScale, color, thickness)

    # Display frames
    cv2.imshow('YouTube Video', img)

    # Break the loop when 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close all windows
cap.release()
cv2.destroyAllWindows()
```

```python

from ultralytics import YOLO
import cv2
import math 
# start webcam
cap = cv2.VideoCapture(0)
cap.set(3, 640)
cap.set(4, 480)

# model
# model = YOLO("yolo-Weights/yolov8n.pt")
model = YOLO('C:/Users/user/Data_Ana_3/Sprint2/Final_files/runs/detect/yolov8sdronebirdlight.pt/weights/best.pt')  # 70 times training
# object classes
classNames = model.names

while True:
    success, img = cap.read()
    results = model(img, stream=True)

    # coordinates
    for r in results:
        boxes = r.boxes

        for box in boxes:
            # bounding box
            x1, y1, x2, y2 = box.xyxy[0]
            x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2) # convert to int values

            # put box in cam
            cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 255), 3)

            # confidence
            confidence = math.ceil((box.conf[0]*100))/100
            print("Confidence --->",confidence)

            # class name
            cls = int(box.cls[0])
            print("Class name -->", classNames[cls])

            # object details
            org = [x1, y1]
            font = cv2.FONT_HERSHEY_SIMPLEX
            fontScale = 1
            color = (255, 0, 0)
            thickness = 2

            cv2.putText(img, classNames[cls], org, font, fontScale, color, thickness)

    cv2.imshow('Webcam', img)
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
